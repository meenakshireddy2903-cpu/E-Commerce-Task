#Analysing Ecommerce Sales Drop

Company Scenario (Flipkart/Amazon):
We are seeing a  drop in revenue find the reason and suggest actions.

```python
#import necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#load the data
customers = pd.read_csv("customers.csv")
orders = pd.read_csv("orders.csv")
products = pd.read_csv("products.csv")
```

**EDA(EXPLORATORY DATA ANALYSIS)**

```python
customers
```

```python
orders
```

**joining**(customers and orders nd products)

```python
df = pd.merge(customers,orders,on="customer_id",how="inner")
df = pd.merge(df,products, on="product_id",how = "inner")
df
```

```python
df.info()
```

```python
#Converting object date columns into date time columns
df["signup_date"] = pd.to_datetime(df["signup_date"])
df["order_date"] = pd.to_datetime(df["order_date"])
df.info()
```

```python
df
```

```python
#Create Revenue column
#Formula: Revenue = Price*Quantity
df["revenue"] = df["price"]*df["quantity"]
df.head()
#revenue column is created
```

```python
#Revenue Trend
revenue = df.groupby("order_date")["revenue"].sum()  #.sort_values() can be used to sort the revenue
revenue = pd.DataFrame(revenue).reset_index()
plt.plot(revenue["order_date"],revenue["revenue"],marker = "o",mfc = "orange",mec = "orange")
plt.xlabel("Order Date")
plt.ylabel("Revenue")
plt.title("Revenue by Order Date")
plt.show()
#revenue
```

**The revenue is very less in the last 2 months(Mid MAY,JUN,JUL)**

Here we came to know in which month sales are dropping, so now checking in which city.

```python
df["city"].unique()
```

```python
#Revenue Trend by city
revenue = df.groupby("city")["revenue"].sum()  #.sort_values() can be used to sort the revenue
revenue = pd.DataFrame(revenue).reset_index()
plt.plot(revenue["city"],revenue["revenue"],marker = "o",mfc = "orange",mec = "orange")
plt.xlabel("city")
plt.ylabel("Revenue")
plt.title("Revenue by city")
plt.show()
```

**Revenue is dropping in Banglore**

```python
#Revenue Trend by category
revenue = df.groupby("category")["revenue"].sum()  #.sort_values() can be used to sort the revenue
revenue = pd.DataFrame(revenue).reset_index()
plt.plot(revenue["category"],revenue["revenue"],marker = "o",mfc = "orange",mec = "orange")
plt.xlabel("category")
plt.ylabel("Revenue")
plt.title("Revenue by category")
plt.show()
```

**In Fashion Category Revenue got dropped**

```python
#Count customers based on order_id(count of orders per customer)
#df["customer_id"].count()
df.groupby("customer_id")["order_id"].count()
```

**Most of the customers are ordering less**

```python
#Revenue Trend by product_name
revenue = df.groupby("product_name")["revenue"].sum()  #.sort_values() can be used to sort the revenue
revenue = pd.DataFrame(revenue).reset_index()
plt.plot(revenue["product_name"],revenue["revenue"],marker = "o",mfc = "orange",mec = "orange")
plt.xlabel("product_name")
plt.ylabel("Revenue")
plt.title("Revenue by product_name")
```

**Revenue is very less for shoes and watch**

#Insights:

1. Revenue got dropped in last two months (jun,jul)

2. Banglore contributed more in revenue drop.

3. Fashion category recorded less revenue.

4. Repeated customers ordered less.

5. Shoe and watch sales declined significantly.


# Recommendations:
1. Marketing should be done in Banglore.

2. Provide discounts for shoe and Watch.

3. Coupons, Discounts, Loyalty Programs for repeated customers.
